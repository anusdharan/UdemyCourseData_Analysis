```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt #basic Vizulization Library 
%matplotlib inline 
```

```python
data=pd.read_csv(r"C:\Users\Admin\Downloads\practise1\Udemy_Courses.csv")
```

```python
#Display Top 10 Rows of The Dataset
data.head(10)
```

```python
#Check the Last 5 Rows of The Dataset
data.tail(5)
```

```python
#Find Shape of Our Dataset (Number of Rows & Number of Columns)
data.shape
```

```python
#Get Information About Our Dataset Like Total Number Rows, 
#Total Number of Columns, Datatypes of Each Column And Memory Requirement
data.info()
```

```python
#Check Null Values In The Dataset
data.isnull().sum()
```

```python
sns.heatmap(data.isnull())
```

```python
#Check For Duplicate Data and Drop Them
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

```python
#Find Out Number of Courses Per Subjects
data.head(2)
```

```python
data['subject'].value_counts()
```

```python
sns.countplot(data['subject'])
plt.xlabel("Subjects",fontsize=13)
plt.ylabel("Number of Courses Per Subject",fontsize=13)
plt.xticks(rotation=65)
plt.show()
```

```python
#For Which Levels, Udemy Courses Providing The Courses
data.head(2)
```

```python
data['level'].value_counts()
```

```python
sns.countplot(data['level'])
plt.xlabel("Level",fontsize=13)
plt.ylabel("Number of Courses Per Level",fontsize=13)
plt.xticks(rotation=65)
plt.show()
```

```python
#Display The Count of Paid and Free Courses
data.head(2)
```

```python
data['is_paid'].value_counts()
```

```python
sns.countplot(data['is_paid'])
plt.xlabel("Level",fontsize=13)
plt.ylabel("Number of Free And Paid Courses",fontsize=13)
plt.xticks(rotation=65)
plt.show()
```

```python
#Which Course Has More Lectures (Free or Paid)?
data.head(2)
```

```python
data.groupby('is_paid').mean()
```

```python
sns.barplot(x="is_paid",y="num_lectures",data=data)
```

```python
#Which Courses Have A Higher Number of Subscribers Free or Paid?
sns.barplot(x="is_paid",y="num_subscribers",data=data)
```

```python
#Which Level Has The Highest Number of Subscribers?
data.groupby('level').mean()['num_subscribers']
```

```python
sns.barplot(x="level",y="num_subscribers",data=data)
```

```python
#Find Most Popular Course Title
data.head(2)
```

```python
data[data['num_subscribers'].max()==data['num_subscribers']]['course_title']
```

```python
#Display 10 Most Popular Courses As Per Number of Subscribers
top_10=data.sort_values(by="num_subscribers",ascending=False).head(10)
top_10
```

```python
sns.barplot(x="num_subscribers",y="course_title",data=top_10)
```

```python
#Find The Course Which Is Having The Highest Number of Reviews.
top=data.sort_values(by="num_reviews",ascending=False).head(10)
top
```

```python
sns.barplot(x="num_reviews",y="course_title",data=top_10)
```

```python
#Does Price Affect the Number of Reviews?
plt.figure(figsize=(15,6))
sns.scatterplot(x="price",y="num_reviews",data=data)
```

```python
#Find Total Number of Courses Related To Python
data[data['course_title'].str.contains("python",case=False)]
```

```python
data[data['course_title'].str.contains("python",case=False)].count()
```

```python
len(data[data['course_title'].str.contains("python",case=False)])
```

```python
#Display 10 Most Popular Python Courses As Per Number of Subscribers
python=data[data['course_title'].str.contains('python',case=False)].sort_values(by="num_subscribers",ascending=False).head(10)
```

```python
python
```

```python
sns.barplot(x="num_subscribers",y="course_title",data=python)
```

```python
 #In Which Year The Highest Number of Courses Were Posted?
data.head(2)
```

```python
data.published_timestamp=pd.to_datetime(data.published_timestamp)
```

```python
data.head(2)
```

```python
data['Year']=data.published_timestamp.dt.year
```

```python
data.insert(1,'year',data.published_timestamp.dt.year)
```

```python
data.head(2)
```

```python
data.groupby('year').count()
```

```python
sns.countplot('Year',data=data)
```

```python
#Display Category-Wise Count of Posted Subjects [Year Wise]
data.groupby('year')['subject'].value_counts()
```

```python
data.groupby('year')['subject'].value_counts().plot(kind="bar")
```
