# 7) Show all courses of Graphic Design where the price is below 100 ?

```python
data['price'].unique()
```

```python
#Some Transformations
data['price'] = data['price'].astype(str) #Changing the price to string
data['price'] = data['price'].apply(lambda x: x.replace('$',''))#Using lambda function to replace ',' from price
data['price'] = data['price'].apply(lambda x: x.replace('Free','0'))
data['price'] = data['price'].astype(float) # Changing the price to Float
data.info() # looking at the dataset information after transformation
```

```python
data[(data['subject']=='Graphic Design')&(data['price']<100)]
```
